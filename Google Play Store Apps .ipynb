```python
 histogram displaying the distribution of app ratings for Google Play Store apps. The x-axis represents the app
ratings, while the y-axis shows the count of apps with each rating. Most app ratings are concentrated around the
4.0 to 5.0 range, with a significant peak near 4.5. There are fewer apps with ratings below 3.0, and very few 
outliers with ratings above 7.5. This indicates that the majority of apps on the platform receive favorable 
ratings, clustering around higher values.
```
